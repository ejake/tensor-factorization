```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Read Input
lines = np.loadtxt("/home/rajaquep/jou/resultsfile.txt", skiprows=1, usecols=(1,2,3,4), unpack=False)
```

```python
import csv
veca = []
vece = []
veci = []
veci2 = []
veco = []
vecu = []
with open('/home/rajaquep/jou/resultsfile.txt', 'rb') as f:
    reader = csv.reader(f, delimiter='\t')    
    for row in reader:
        if row[1]=='1':
            veca.append( map(float, row[2:6]) )
        if row[1]=='2':
            vece.append( map(float, row[2:6]) )
        if row[1]=='3':
            veci.append( map(float, row[2:6]) )
        if row[1]=='4':
            veci2.append( map(float, row[2:6]) )
        if row[1]=='5':
            veco.append( map(float, row[2:6]) )
        if row[1]=='6':
            vecu.append( map(float, row[2:6]) )
```

```python
np.where(lines[:,0]==1)
```

```python
plt.plot( lines[np.where(lines[:,0]==1),2] , lines[np.where(lines[:,0]==1),3],marker=r'$a$', c='b',label=r'$a$')
plt.plot( lines[np.where(lines[:,0]==2),2] , lines[np.where(lines[:,0]==2),3],marker=r'$e$', c='r',label=r'$e$')
plt.plot( lines[np.where(lines[:,0]==3),2] , lines[np.where(lines[:,0]==3),3],marker=r'$i$', c='g',label=r'$i$')
plt.plot( lines[np.where(lines[:,0]==4),2] , lines[np.where(lines[:,0]==4),3],marker=r'$\hat{i}$', c='c',label=r'$\hat{i}$')
plt.plot( lines[np.where(lines[:,0]==5),2] , lines[np.where(lines[:,0]==5),3],marker=r'$o$', c='y',label=r'$o$')
plt.plot( lines[np.where(lines[:,0]==6),2] , lines[np.where(lines[:,0]==6),3],marker=r'$u$', c='m',label=r'$u$')
```

```python
plt.plot( np.asarray(veca)[:,2] , np.asarray(veca)[:,3],marker=r'$a$', c='b',label=r'$a$',linestyle='None',markersize=8)
plt.plot( np.asarray(vece)[:,2] , np.asarray(vece)[:,3],marker=r'$e$', c='r',label=r'$e$',linestyle='None',markersize=8)
plt.plot( np.asarray(veci)[:,2] , np.asarray(veci)[:,3],marker=r'$i$', c='g',label=r'$i$',linestyle='None',markersize=10)
plt.plot( np.asarray(veci2)[:,2] , np.asarray(veci2)[:,3],marker=r'$\hat{i}$', c='c',label=r'$\i$',linestyle='None', markersize=10)
plt.plot( np.asarray(veco)[:,2] , np.asarray(veco)[:,3],marker=r'$o$', c='y',label=r'$o$',linestyle='None',markersize=8)
plt.plot( np.asarray(vecu)[:,2] , np.asarray(vecu)[:,3],marker=r'$u$', c='m',label=r'$u$',linestyle='None',markersize=8)
plt.xlabel('F1')
plt.ylabel('F2')
```
